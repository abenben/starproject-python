## 7.5: グラフのスタイルと書式の設定

```python
# リスト 7.5.1　スタイル "whitegrid" の散布図を描画

%matplotlib inline
import numpy as np
import seaborn as sns

np.random.seed(1)
sns.set_style(style="whitegrid")
x, y = np.random.rand(2, 100)
sns.scatterplot(x=x, y=y)
```

```python
# リスト 7.5.2　matplotlib.rcParams を上書きして描画
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1, 2], [1, 2])
```

```python
# リスト 7.5.3　スタイルが反映されたグラフ

import pandas as pd

np.random.seed(1)
ser = pd.Series(np.random.rand(50).cumsum())
ser.plot()
```

```python
# リスト 7.5.4　グラフのスタイルを初期化して描画

sns.reset_defaults()
ser.plot()
```

```python
# リスト 7.5.5　サブプロットの上側と右側の枠線を非表示に設定

sns.set_style(style={"axes.spines.top": False, "axes.spines.right": False})
# 下記と同じ
# sns.set_style(rc={"axes.spines.top": False, "axes.spines.right": False})
ser.plot()
```

```python
# リスト 7.5.6　スタイルと書式を設定したグラフ

sns.reset_defaults()
sns.set_style(style="darkgrid", rc={"grid.linestyle": "--"})
ser.plot()
```

```python
# リスト 7.5.7　seaborn で設定している書式を確認

sns.axes_style()
```

```python
# リスト 7.5.8　seaborn と Matplotlib を併用した書式設定

sns.reset_defaults()
sns.set_style({"axes.facecolor": "#eee"})
ax = sns.scatterplot(x=x, y=y)
ax.grid(which="major", axis="y", linestyle="--", linewidth=3)
```
