## 3.4: カテゴリデータの可視化

```python
# 3.4.1　tips データセットの読み込み

import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
# 3.4.2　total_bill の値をストリッププロットに描画

sns.catplot(data=tips, x="time", y="total_bill", kind="strip")
```

```python
# 3.4.3　smoker 列で色分けしたストリッププロット

sns.catplot(data=tips, x="total_bill", y="time", hue="smoker")
```

```python
# 3.4.4　ジッタをオフにしたストリッププロット

sns.catplot(data=tips, x="time", y="total_bill", jitter=0)
```

```python
# 3.4.5　ビースウォームプロットの描画

sns.catplot(data=tips, x="time", y="total_bill", hue="smoker", kind="swarm")
```

```python
# 3.4.6　箱ひげ図の描画

sns.catplot(data=tips, x="time", y="total_bill", kind="box")
```

```python
# 3.4.7　箱ひげ図を smoker 列で分割して描画

sns.catplot(data=tips, x="time", y="total_bill", kind="box", hue="smoker")
```

```python
# 3.4.8　分位数を増やした箱ひげ図

sns.catplot(data=tips, x="time", y="total_bill", kind="boxen", hue="smoker")
```

```python
# 3.4.9　バイオリン図の描画

sns.catplot(data=tips, x="time", y="total_bill", kind="violin")
```

```python
# 3.4.10　左右に分割したバイオリン図

sns.catplot(
    data=tips, x="time", y="total_bill", kind="violin", hue="smoker", split=True
)
```

```python
# 3.4.11　棒グラフの描画

sns.catplot(data=tips, x="day", y="total_bill", kind="bar")
```

```python
# 3.4.12　中央値で集計した棒グラフの描画

import numpy as np

sns.catplot(
    data=tips, x="day", y="total_bill", kind="bar", hue="smoker", estimator=np.median
)
```

```python
# 3.4.13　データの個数を棒グラフに描画

sns.catplot(data=tips, x="day", kind="count")
```

```python
# 3.4.14　ポイントプロットを描画

sns.catplot(data=tips, x="day", y="total_bill", kind="point")
```

```python
# 3.4.15　推定値を中央値にして sex 列で分類したポイントプロット

sns.catplot(
    data=tips, x="day", y="total_bill", kind="point", estimator=np.median, hue="sex"
)
```

```python
# 3.4.16　smoker 列を縦に、sex 列を横に並べた箱ひげ図

sns.catplot(data=tips, x="day", y="total_bill", kind="box", row="smoker", col="sex")
```
