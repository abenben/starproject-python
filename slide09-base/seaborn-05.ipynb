## 6.6: データ分布の可視化

```python
# リスト 6.6.1　tips データセットの読み込み

import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
# リスト 6.6.2　ヒストグラムとカーネル密度推定グラフの描画

sns.distplot(tips["total_bill"])
```

```python
# リスト 6.6.3　ヒストグラムとラグプロットの描画

sns.distplot(tips["total_bill"], hist=True, kde=False, rug=True, bins=10)
```

```python
# リスト 6.6.4　度数と相対度数のヒストグラムを並べて描画

import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(figsize=(10, 4), nrows=1, ncols=2)
ax1.set_title("norm_hist=False")
ax2.set_title("norm_hist=True")
sns.distplot(tips["total_bill"], kde=False, ax=ax1)
sns.distplot(tips["total_bill"], kde=False, norm_hist=True, ax=ax2)
```

```python
# リスト 6.6.5　正規分布とガンマ分布を適合させたヒストグラム

from scipy import stats

fig, ax = plt.subplots()
sns.distplot(
    tips["total_bill"],
    kde=False,
    fit=stats.norm,  # ① 正規分布
    fit_kws={"label": "norm", "color": "g"},  # ③ 線のラベルと書式
    ax=ax,
)
sns.distplot(
    tips["total_bill"],
    kde=False,
    fit=stats.gamma,  # ② ガンマ分布
    fit_kws={"label": "gamma", "color": "k"},  # ③ 線のラベルと書式
    ax=ax,
)
ax.legend()
```

```python
# リスト 6.6.6　散布図とヒストグラム

sns.jointplot(data=tips, x="tip", y="total_bill")
```

```python
# リスト 6.6.7　カーネル密度推定をしたグラフ

sns.jointplot(data=tips, x="tip", y="total_bill", kind="kde")
```
