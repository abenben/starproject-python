## 6.3: グラフの描画方法

```python
# 6.3.1　tips データセットの読み込み

import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
# リスト 6.3.2　scatterplot() 関数による散布図の描画

sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# リスト 6.3.3　Matplotlib で配置やグラフタイトルを設定

import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax1)
sns.scatterplot(data=tips, x="size", y="tip", ax=ax2)
ax1.set_title("totall bill and tips")
ax2.set_title("size and tips")
```

```python
# リスト 6.3.4　散布図とヒストグラムを並べて描画

sns.jointplot(data=tips, x="total_bill", y="tip")
```

```python
# リスト 6.3.5　smoker 列でデータを分割して描画

sns.relplot(data=tips, x="total_bill", y="tip", row="smoker")
```

```python
# リスト 6.3.6　time 列でデータを分割して描画

sns.relplot(data=tips, x="total_bill", y="tip", col="time")
```

```python
# リスト 6.3.7　smoker 列と time 列でデータを分割して描画

sns.relplot(data=tips, x="total_bill", y="tip", row="smoker", col="time")
```

```python
# リスト 6.3.8　pairplot() 関数を用いて散布図行列を描画

sns.pairplot(data=tips, vars=["total_bill", "tip", "size"])
```

```python
# リスト 6.3.9　雑然データの DataFrame を作成

import pandas as pd

wide_df = pd.DataFrame(
    [["Tokyo", 5.8, 5.7, 5.6], ["Osaka", 6.8, 6.7, 6.6], ["Nagoya", 5.1, 5.1, 5.0],],
    columns=["Location", "2020-01-01", "2020-01-02", "2020-01-03"],
)
wide_df
```

```python
# リスト 6.3.10　melt メソッドによる整然データへの変換

tidy_df = wide_df.melt(id_vars="Location", var_name="Date", value_name="Temperature")
tidy_df
```

```python
# リスト 6.3.11　catplot() 関数を利用した棒グラフの描画

sns.catplot(data=tidy_df, x="Date", y="Temperature", kind="bar", hue="Location")
```
