```python
# リスト 4.2.4　add_axes() メソッドでサブプロットを配置

fig = plt.figure(linewidth=5, edgecolor="k")

# left, bottom, width, height
rect1 = 0.2, 0.2, 0.3, 0.3
rect2 = 0.5, 0.5, 0.4, 0.4

# サブプロットオブジェクトを配置
ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
fig.savefig("add_axes.png", linewidth=5, edgecolor="k")
```

```python
# リスト 4.2.6　add_subplots() メソッドによるサブプロットの配置

# フィギュアオブジェクトの生成
fig = plt.figure()

# フィギュア内にサブプロットを3つ配置
ax1 = fig.add_subplot(2, 2, 1)  # 2行2列の1番
ax2 = fig.add_subplot(2, 2, 2)  # 2行2列の2番
ax3 = fig.add_subplot(2, 2, 3)  # 2行2列の3番

# サブプロット名とサブプロット番号をサブプロット内に描画
for i, ax in enumerate([ax1, ax2, ax3], start=1):
    txt = f"ax{i}\n(2, 2, {i})"
    ax.text(0.2, 0.5, txt)
```

```python
# リスト 4.2.7　subplots() メソッドによるサブプロットの配置

import itertools

# フィギュアオブジェクトの生成
fig = plt.figure()

# subplots()による2行3列のサブプロットオブジェクトの生成
lines, columns = 2, 3
axes = fig.subplots(lines, columns)

# サブプロットの位置をサブプロット内に描画
# 行番号を「l」、列番号を「c」で指定
for l, c in itertools.product(range(lines), range(columns)):
    axes[l, c].text(0.1, 0.5, f"axes[{l}, {c}]")
```
