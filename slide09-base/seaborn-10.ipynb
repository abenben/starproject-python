## 3.10: axes-level 関数の利用(seaborn)

```python
import seaborn as sns

import warnings
warnings.filterwarnings("ignore", module="seaborn")
```

```python
# 3.10.1　箱ひげ図とラグプロットを重ねて描画

tips = sns.load_dataset("tips")
sns.boxplot(data=tips, x="total_bill")
sns.rugplot(tips["total_bill"])
```

```python
# 3.10.2　4 種類のグラフを各サブプロットに描画

import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
sns.regplot(data=tips, x="total_bill", y="tip", ax=ax[0][0])
sns.violinplot(data=tips, x="time", y="tip", hue="smoker", split=True, ax=ax[0][1])
sns.distplot(tips["tip"], ax=ax[1][0])
sns.kdeplot(tips["total_bill"], tips["tip"], ax=ax[1][1])
```

```python
# 3.10.3　散布図の描画

sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# 3.10.4　要素を分割した散布図を描画

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker", style="time")
```

```python
# 3.10.5　折れ線グラフの描画

fmri = sns.load_dataset("fmri")
sns.lineplot(data=fmri, x="timepoint", y="signal")
```

```python
# 3.10.6　折れ線グラフを色分けして描画

import numpy as np

sns.lineplot(
    data=fmri,
    x="timepoint",
    y="signal",
    estimator=np.median,
    hue="event",
    style="event",
    err_style="bars",
    ci=68,
)
```

```python
# 3.10.7　ストリッププロットの描画

sns.stripplot(data=tips, x="smoker", y="tip")
```

```python
# 3.10.8　カスタマイズしたストリッププロットを描画

sns.stripplot(
    data=tips, x="tip", y="day", orient="h", hue="smoker", jitter=0, dodge=True
)
```

```python
# 3.10.9　ビースウォームプロットの描画

sns.swarmplot(data=tips, x="time", y="tip")
```

```python
# 3.10.10　カスタマイズしたビースウォームプロットを描画

sns.swarmplot(data=tips, x="tip", y="day", orient="h", hue="smoker", dodge=True)
```

```python
# 3.10.11　箱ひげ図の描画

sns.boxplot(data=tips, x="time", y="tip")
```

```python
# 3.10.12　箱ひげ図を色分けして横向きに描画

sns.boxplot(data=tips, x="tip", y="day", orient="h", hue="smoker")
```

```python
# 3.10.13　分位数を増やした箱ひげ図の描画

sns.boxenplot(data=tips, x="day", y="tip", hue="smoker")
```

```python
# 3.10.14　バイオリン図の描画

sns.violinplot(data=tips, x="time", y="tip")
```

```python
# 3.10.15　バイオリン図を色分けして描画
sns.violinplot(
    data=tips, x="tip", y="day", orient="h", hue="smoker", split=True, inner="quartile"
)
```

```python
# 3.10.15　バイオリン図を色分けして描画

sns.violinplot(
    data=tips, x="tip", y="day", orient="h", hue="smoker", split=True, inner="quartile"
)
```

```python
# 3.10.17　ポイントプロットを色分けして描画

import numpy as np

sns.pointplot(
    data=tips, x="day", y="tip", hue="smoker", ci=66, dodge=True, estimator=np.median
)
```

```python
# 3.10.18　棒グラフの描画

sns.barplot(data=tips, x="day", y="tip")
```

```python
# 3.10.19　棒グラフを色分けして描画

sns.barplot(
    data=tips, x="tip", y="day", orient="h", hue="smoker", estimator=np.median, ci=68
)
```

```python
# 3.10.20　データの個数を棒グラフに描画

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
sns.countplot(data=tips, x="day", ax=ax[0])
sns.countplot(data=tips, y="smoker", ax=ax[1])
```

```python
# 3.10.21　データの個数をグループ化して棒グラフに描画

sns.countplot(data=tips, x="day", hue="smoker")
```

```python
# 3.10.22　ヒストグラムとカーネル密度推定グラフを重ねて描画

sns.distplot(tips["tip"])
```

```python
# 3.10.23　distplot() に描画方法を指定

sns.distplot(tips["tip"], vertical=True, hist=True, kde=True, rug=True)
```

```python
# 3.10.24　カーネル密度推定グラフの描画

sns.kdeplot(tips["tip"])
```

```python
# 3.10.25　2 変数の分布を等高線として描画

sns.kdeplot(data=tips["total_bill"], data2=tips["tip"])
```

```python
# 3.10.26　ラグプロットを描画

sns.rugplot(tips["tip"])
```

```python
# 3.10.27　Y 軸に沿って描画

sns.rugplot(tips["tip"], axis="y")
```

```python
# 3.10.28　回帰グラフの描画

sns.regplot(data=tips, x="total_bill", y="tip")
```

```python
# 3.10.29　信頼区間を変更した回帰グラフを描画

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
sns.regplot(data=tips, x="size", y="tip", ax=ax1)
sns.regplot(data=tips, x="size", y="tip", ci=68, x_jitter=0.2, ax=ax2)
```

```python
# 3.10.30　代表値を点で描画

sns.regplot(data=tips, x="size", y="tip", x_estimator=np.median)
```

```python
# 3.10.31　回帰線からの残差を描画

sns.residplot(data=tips, x="total_bill", y="tip")
```

```python
# 3.10.32　Lowess 平滑化とロバスト回帰モデルを利用

sns.residplot(data=tips, x="total_bill", y="tip", lowess=True, robust=True)
```

```python
# 3.10.33　flights データセットの読み込み

flights = sns.load_dataset("flights").pivot("month", "year", "passengers")
flights.head()
```

```python
# 3.10.34　ヒートマップの描画

sns.heatmap(flights)
```

```python
# 3.10.35　指定したカラーマップでヒートマップを描画

sns.heatmap(flights, cmap="YlGnBu", annot=True, fmt="d")
```
